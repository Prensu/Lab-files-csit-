## Program to implement the Multiple Linear Regression algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# New dataset: House Size vs Price (in $1000s)
data = {
    'House_Size_sqft': [800, 1000, 1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600],
    'House_Price_1000s': [150, 180, 200, 230, 260, 280, 310, 330, 360, 390]
}
df = pd.DataFrame(data)

# Features and target
X = df[['House_Size_sqft']]  # 2D array
y = df['House_Price_1000s']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Output results
print("üéØ Coefficient (Slope):", model.coef_[0])
print("üìç Intercept:", model.intercept_)
print("\nüìà Predicted values:", y_pred)
print("‚úÖ R2 Score:", r2_score(y_test, y_pred))
print("üîç Mean Squared Error:", mean_squared_error(y_test, y_pred))

# Visualize
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($1000s)")
plt.title("Simple Linear Regression: House Price Prediction")
plt.legend()
plt.grid(True)
plt.show()

# Predict for new value
sample = [[2100]]  # e.g., 2100 sq ft house
predicted_price = model.predict(sample)
print("\nüìå Predicted price for 2100 sq ft house ($1000s):", round(predicted_price[0], 2))
```
